### Data wrangling

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df_case = pd.read_excel('Case_1_20230123.xlsx')
df_case.head()
```

```python
df = df_case.copy()
df.head()
```

```python
df.drop('Nome', axis=1, inplace=True)
df.head()
```

```python
base_color = sns.color_palette()[0]
```

### Análises univariadas

Para departamento:

```python
dpt_order = df['Departamento'].value_counts().index

plt.figure(figsize = [15, 10])
ax = sns.countplot(data=df, y='Departamento', color=base_color, order=dpt_order)

ax.bar_label(ax.containers[0])
plt.title('Número de funcionários por departamento')
plt.xlabel('Frequência');
```

```python
dpt_counts = df['Departamento'].value_counts() # frequencia dos valores unicos da coluna 

n_dpt = df['Departamento'].value_counts().sum() # soma de todos os valores não-nulos da coluna

max_dpt_count = dpt_counts[0] # maior frequencia da coluna

max_prop = max_dpt_count / n_dpt # maior proporcao

tick_props = np.arange(0, max_prop, 0.05) # Criar uma array com valores espaçados igualmente proporcionais

tick_names = ['{:0.2f}'.format(v) for v in tick_props] # Criar uma lista de valores de string para serem usados como rótulos dos tick

plt.figure(figsize = [15, 10])
sns.countplot(data=df, y='Departamento', color=base_color, order=dpt_order)

plt.xticks(tick_props * n_dpt, tick_names)
plt.xlabel('Proporção de cada departamento')

for i in range (dpt_counts.shape[0]):
    count = dpt_counts[i]
    pct_string = '{:0.1f}%'.format(100*count/n_dpt)
    plt.text(count+1, i, pct_string, va='center');
```
