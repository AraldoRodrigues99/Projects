Import the necessary modules:

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

Run the dataset in the notebook:

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

Check if there are null values:

```python
pd.isnull(df_tracks).sum()
```

Get a general insight about the dataset:

```python
df_tracks.info()
```

See the 10 least popular musics:

```python
sorted_df = df_tracks.sort_values('popularity', ascending=True).head()
sorted_df
```

See the descriptive statistics of the numerical variables:

```python
df_tracks.describe().transpose()
```

See the 10 most popular songs:

```python
most_popular = df_tracks.query('popularity > 90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

Set the release date column to be the index of the dataframe and change its datatype:

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

Check the artist of a determined row:

```python
df_tracks[['artists']].iloc[18]
```

Convert the duration_ms to seconds, and drop it:

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x / 1000))
df_tracks.drop('duration_ms', inplace=True, axis = 1)
df_tracks['duration'].head()
```

Create a heatmap with the correlation of the variables:

```python
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')

plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

Get a sample of 4% of the original dataframe:

```python
sample_df = df_tracks.sample(int(0.004 * len(df_tracks))) 
len(sample_df)
```

Plot the relation between loudness and correlation:

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation');
```

Plot the relation between popularity and acousticness:

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation');
```

Create another column in the dataframe with the release date of each song and plot its year distribution:

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date') # get the index values to the new column
df_tracks['dates'] = pd.to_datetime(df_tracks['dates']) # change datatype
years = df_tracks['dates'].dt.year # get the year of each date
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year');
```

Plot the duration of songs over the years:

```python
total_dr = df_tracks['duration']
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title='Years vs Duration')
plt.xticks(rotation=90);
```

Load the other dataset:

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

Plot the distribution of duration for each gender:

```python
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
plt.title('Duration of the Songs in Different Genres')
```

Plot the popularity of the genres of the top 10 most popular songs:

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10, 5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title='Top 5 Genres by Popularity');
```
